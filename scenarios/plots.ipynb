```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
means = {}
for model in ['FIFO', 'PPOPlanner_interval', 'PPOPlanner', 'Random', 'ShortestProcessingTime']:
    df = pd.read_csv(f'./results/{model}_slow_server.txt')
    means[model] = df.mean_cycle_time.mean()
    

courses = list(means.keys())
values = list(means.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='maroon', 
        width = 0.4)
 
plt.xlabel("Model. In PPOPlanner_interval, the model is evaluated in intervals.")
plt.ylabel("Mean cycle time")
plt.title("Mean cycle time per model for the slow_server model")
plt.show()
```
